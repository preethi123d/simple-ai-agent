```python
memory = []

def add_to_memory(user, response):
    memory.append({"user": user, "agent": response})
```

```python
def planner(user_input):
    text = user_input.lower()

    if "exam" in text:
        return "study_plan"
    elif "deadline" in text:
        return "task_management"
    elif "dsa" in text or "algorithm" in text:
        return "dsa_help"
    else:
        return "general"
```

```python
def tool_selector(intent):
    tools = {
        "study_plan": "make_plan",
        "task_management": "prioritize",
        "dsa_help": "suggest_topics",
        "general": "basic_chat"
    }
    return tools[intent]
```

```python
import re

def executor(tool, user_input):

    text = user_input.lower()

    # Extract number of days (default = 3)
    days = 3
    match = re.search(r'\d+', text)
    if match:
        days = int(match.group())

    if tool == "make_plan":

        if days == 1:
            return """⚡ 1-Day Emergency Plan:
- Revise Arrays & Strings (2 hrs)
- Practice 5 easy problems
- Watch 1 quick revision video
- Sleep well before exam"""

        elif days == 2:
            return """📅 2-Day Plan:
Day 1: Arrays, Strings
Day 2: Searching + Practice"""

        elif days >= 3:
            plan = "📅 Study Plan:\n"
            topics = ["Arrays", "Strings", "Searching", "Recursion", "DP"]

            for i in range(days):
                topic = topics[i % len(topics)]
                plan += f"Day {i+1}: {topic}\n"

            return plan

    elif tool == "prioritize":
        return """📝 Task Strategy:
1. Urgent tasks first
2. Break into small steps
3. Avoid multitasking"""

    elif tool == "suggest_topics":
        return """💡 Focus on:
- Arrays
- Strings
- Binary Search
- Recursion"""

    else:
        return "Tell me more, I’ll help you plan 😊"
```

```python
def agent(user_input):

    intent = planner(user_input)
    tool = tool_selector(intent)
    result = executor(tool, user_input)

    add_to_memory(user_input, result)

    return {
        "intent": intent,
        "tool_used": tool,
        "response": result
    }
```

```python
user_input = "I have exam in 3 days and didn't study DSA"

output = agent(user_input)

print("Intent:", output["intent"])
print("Tool Used:", output["tool_used"])
print("\nResponse:\n", output["response"])
```

```python
while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        break

    output = agent(user_input)

    print("\nAgent:")
    print("Intent:", output["intent"])
    print("Tool:", output["tool_used"])
    print(output["response"])
    print("-" * 40)
```
